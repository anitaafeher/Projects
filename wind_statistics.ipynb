### Step 1. Import the necessary libraries

`````python
import pandas as pd
import numpy as np
`````

Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

`````python
data = pd.read_csv('wind.csv')
df = pd.DataFrame(data)

df['Date'] = df.apply(lambda x: f"{int(x['Yr'])}-{int(x['Mo'])}-{int(x['Dy'])}", axis=1) #strings together the first 3 cols
df['Date'] = pd.to_datetime(df['Date'], format='%y-%m-%d') #converts to date
df = df.drop(['Yr', 'Mo', 'Dy'], axis=1) #drops the first 3 cols 
first_col = df.pop('Date')
df.insert(0, 'Date', first_col) #makes Date the first column
df
`````

### Step 2. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

`````python
def year_correction(date):
    if date.year > 1989:
        date = date.replace(year=date.year -100)
    return date

wrong_dates = df['Date'].apply(lambda x: x.year > 1989).sum()
print(f"Count of the wrong dates: {wrong_dates}")
df['Date'] = df['Date'].apply(year_correction)
print(df)
`````

### Step 3. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

`````python
df['Date'] = pd.to_datetime(df['Date']) #making sure it is date type
df.set_index('Date', inplace=True)
print(df)
`````

### Step 4. Compute how many values are missing for each location over the entire record.  
#### They should be ignored in all calculations below. 

`````python
nan_values = df.isna().sum()
print(f"Summary of the missing values by location:\n{nan_values}")
`````

### Step 5. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
#### A single number for the entire dataset.

`````python
average_windspeed = df.stack().mean()
rounded_average = round(average_windspeed, 2)
print(f"Average windspeed over all locations: {rounded_average}")
`````

### Step 8. Find the average windspeed in January for each location.  
#### Treat January 1961 and January 1962 both as January.

`````python
#ensuring the datetime format 
df.index = pd.to_datetime(df.index)

january_data = df[df.index.month == 1]
average_january_windspeed = january_data.mean()

average_january_windspeed
`````

### Step 9. Downsample the record to a yearly frequency for each location.
